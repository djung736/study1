# 1. 프로젝트 load_digits()

## 학습 요구사항

1. Feature Data 지정하기
2. Label Data 지정하기
3. Target Names 출력해 보기
4. 데이터 Describe 해 보기
5. train, test 데이터 분리
6. 다양한 모델로 학습시켜보기

7. 학습데이터 X_train, y_train 을 활용해 분류기 모델을 만들어 봅시다.    
어떤 모델이 가장 좋은 성능을 보일까요?   
     Decision Tree 사용해 보기   
     Random Forest 사용해 보기   
     SVM 사용해 보기   
     SGD Classifier 사용해 보기   
     Logistic Regression 사용해 보기   
8. 모델을 평가해 보기

학습된 모델들의 테스트데이터 예측 결과를 어떻게 해석해야 할까요? 모델의 성능을 평가하는 지표로는 무엇이 좋을까요? sklearn.metrics 에서 제공하는 평가지표 중 적절한 것을 선택해 보세요. 선택하신 이유도 설명해 주세요.

```python
# 함수 불러오기
from sklearn.datasets import load_digits
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 손글씨 불러오기
digits = load_digits() 
digits_data = digits.data #1. Feature Data 지정하기
digits_label = digits.target #2. Label Data 지정하기
print(digits.target_names) #3. Target Names 출력해 보기
print(digits.DESCR) #4. 데이터 Describe 해 보기


# 5. train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size=0.2, 
                                                    random_state=7)

# 여러 모델 생성
decision_tree = DecisionTreeClassifier(random_state=32)
random_forest = RandomForestClassifier(random_state=32)
svm_model = svm.SVC()
sgd_model = SGDClassifier()
logistic_model = LogisticRegression(max_iter=10000) #기범님께 배움 iter 한도 늘리기 워닝 해결

#6 & 7. 다양한 모델로 학습시켜보고 비교
all = [decision_tree,random_forest,svm_model,sgd_model,logistic_model]
acu = 0.0
for i in all:
    i.fit(X_train, y_train) #학습
    y_pred = i.predict(X_test) #테스트에 예측치
    if acu < accuracy_score(y_test,y_pred):
        acu = accuracy_score(y_test,y_pred)
        acc = i
    else:
        pass
    print(i,'\n', classification_report(y_test, y_pred)) #모델 이름 표시 및 줄바꿈으로 가독성향상
print("accuracy가 가장 높은 모델은",acc)
```

* SVC가 0.99로 가장 좋은 결과를 나타냄
* 모델평가성능지표 = Precision을 택했습니다. 검출 못하면 미확인 데이터 처리하면 되지만   
    양성을 음성으로 판단하는게 더 치명적일것 같아서요

어떤 모델이 가장 좋은 성능을 보일까요?   
     Decision Tree 사용해 보기   
     Random Forest 사용해 보기   
     SVM 사용해 보기   
     SGD Classifier 사용해 보기   
     Logistic Regression 사용해 보기   
8. 모델을 평가해 보기

학습된 모델들의 테스트데이터 예측 결과를 어떻게 해석해야 할까요? 모델의 성능을 평가하는 지표로는 무엇이 좋을까요? sklearn.metrics 에서 제공하는 평가지표 중 적절한 것을 선택해 보세요. 선택하신 이유도 설명해 주세요.

```python
# 함수 불러오기
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 와인 불러오기
wine = load_wine() 
wine_data = wine.data #1. Feature Data 지정하기
wine_label = wine.target #2. Label Data 지정하기
print(wine.target_names) #3. Target Names 출력해 보기
print(wine.DESCR) #4. 데이터 Describe 해 보기


# 5. train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(wine_data, 
                                                    wine_label, 
                                                    test_size=0.2, 
                                                    random_state=7657)

# 여러 모델 생성
decision_tree = DecisionTreeClassifier(random_state=16)
random_forest = RandomForestClassifier(random_state=16)
svm_model = svm.SVC()
sgd_model = SGDClassifier()
logistic_model = LogisticRegression(max_iter=5000) #기범님께 배움 iter 한도 늘리기 워닝 해결

#6 & 7. 다양한 모델로 학습시켜보고 비교
all = [decision_tree,random_forest,svm_model,sgd_model,logistic_model]
acu = 0.0
for i in all:
    i.fit(X_train, y_train) #학습
    y_pred = i.predict(X_test) #테스트에 예측치
    if acu < accuracy_score(y_test,y_pred):
        acu = accuracy_score(y_test,y_pred)
        acc = i
    else:
        pass
    print(i,'\n', classification_report(y_test, y_pred,zero_division=1,)) #모델 이름 표시 및 줄바꿈으로 가독성향상
print("accuracy가 가장 높은 모델은",acc)
```

* RandomForest가 0.97로 가장 좋은 결과를 나타냄(사실 1.0나왔지만 만점은 불안해서 랜덤포레스트 랜덤값을 16으로 수정하고 테스트스플릿도 막 변경함)    
(나중에 추가된 에큐스코어로 비교해서 같은 값이지만 나중에 측정된 선형회귀가 무시됨)
* 모델평가성능지표 = accuracy인거 같습니다. 둘다 평균적으로 좋은게 적당할 거 같아서요

# 3. 프로젝트 load_breast_cancer()

```python
# 함수 불러오기
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 유방암 불러오기
breast_cancer = load_breast_cancer() 
breast_cancer_data = breast_cancer.data #1. Feature Data 지정하기
breast_cancer_label = breast_cancer.target #2. Label Data 지정하기
print(breast_cancer.target_names) #3. Target Names 출력해 보기
print(breast_cancer.DESCR) #4. 데이터 Describe 해 보기


# 5. train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data, 
                                                    breast_cancer_label, 
                                                    test_size=0.2, 
                                                    random_state=7657)

# 여러 모델 생성
decision_tree = DecisionTreeClassifier(random_state=16)
random_forest = RandomForestClassifier(random_state=16)
svm_model = svm.SVC()
sgd_model = SGDClassifier()
logistic_model = LogisticRegression(max_iter=10000) #기범님께 배움 iter 한도 늘리기 워닝 해결

#6 & 7. 다양한 모델로 학습시켜보고 비교
all = [decision_tree,random_forest,svm_model,sgd_model,logistic_model]
acu = 0.0
for i in all:
    i.fit(X_train, y_train) #학습
    y_pred = i.predict(X_test) #테스트에 예측치
    if acu < accuracy_score(y_test,y_pred):
        acu = accuracy_score(y_test,y_pred)
        acc = i #큰값저장
    else:
        pass
    print(i,'\n', classification_report(y_test, y_pred)) #모델 이름 표시 및 줄바꿈으로 가독성향상
print("accuracy가 가장 높은 모델은",acc)
```

* accuracy는 randomforest가 0.96으로 제일 높습니다
* 모델평가성능지표 = recall 유방암 판정은 0번(malignant)악성종양을 있는데 없는걸로 판별하면 안되기 때문에 randomforest가 적합한거 같습니다
